```python
import numpy as np
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel("AirBNB.xlsx")
```

```python
df.head(5)
```

```python
sns.countplot(df["room type"] , color="green" )
```

```python
sns.histplot(df['price'], bins=50, kde=True)
plt.title("Distribution of Airbnb Prices")
plt.xlabel("Price ($)")
plt.ylabel("Count")
plt.show()
```

```python
sns.barplot(data=df, x='room type', y='price', estimator='mean', ci=None , palette="rainbow")
plt.title("Average Price by Room Type")
plt.xlabel("Room Type")
plt.ylabel("Average Price ($)")
plt.show()
```

```python
sns.countplot(data=df, y='neighbourhood group', order=df['neighbourhood group'].value_counts().index , color="red")
plt.title("Number of Listings by Neighbourhood Group")
plt.xlabel("Count")
plt.ylabel("Neighbourhood Group")
plt.show()
```

```python
sns.scatterplot(data=df, x='minimum nights', y='price', alpha=0.5)
plt.title("Price vs Minimum Nights Required")
plt.xlabel("Minimum Nights")
plt.ylabel("Price ($)")
plt.xlim(0, 30)
plt.show()
```

```python
corr = df.select_dtypes(include='number').corr()
plt.figure(figsize=(3, 3))

sns.heatmap(corr, cmap='coolwarm', annot=False)
plt.title("Correlation Heatmap of Numeric Features")
plt.show()
```

```python

verified_counts = df['room type'].value_counts()
plt.figure(figsize=(7, 7))
plt.pie(verified_counts, labels=verified_counts.index, autopct='%1.1f%%', startangle=90, colors=sns.color_palette("coolwarm"))
plt.title("Room Type Share")
plt.show()
```

```python
sns.barplot(data=df, x='neighbourhood group', y='review rate number', estimator='mean', ci=None)
plt.title("Average Review Rate by Neighbourhood Group")
plt.xlabel("Neighbourhood Group")
plt.ylabel("Average Review Rate")
plt.show()
```
